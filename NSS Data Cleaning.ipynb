```python
df_bihar_demo = df_bihar[['District','State_region','Sector','hh_size','HH_Type',
                        'Religion','Social_Group','Whether_Land_Owned',
                        'Type_of_land_owned','Land_Owned',
                        'During_july10_june11_cultivated',
                        'During_july10_june11_irrigated']]
df_bihar_demo
```

```python
# map districts
district_map = {'01':'Pashchim Champaran','02':'Purba Champaran','03':'Sheohar','04':'Sitamarhi','05':'Madhubani','06':'Supaul',
             '07':'Araria','08':'Kishanganj','09':'Purnia','10':'Katihar','11':'Madhepura','12':'Saharsa',
             '13':'Darbhanga','14':'Muzaffarpur','15':'Gopalganj','16':'Siwan','17':'Saran','18':'Vaishali',
             '19':'Samastipur','20':'Begusarai','21':'Khagaria','22':'Bhagalpur','23':'Banka','24':'Munger',
             '25':'Lakhisarai','26':'Sheikhpura','27':'Nalanda','28':'Patna','29':'Bhojpur','30':'Buxar','31':'Kaimur (Bhabua)',
             '32':'Rohtas','33':'Aurangabad','34':'Gaya','35':'Nawada','36':'Jamui','37':'Jehanabad','38':'Arwal'}
df_bihar_demo['District_map'] = df_bihar_demo['District'].map(district_map)
df_bihar_demo.head()
df_bihar_demo = df_bihar_demo.drop(['District'], axis=1)
df_bihar_demo.rename({'District_map': 'District'}, axis=1, inplace=True)
df_bihar_demo.head(100)
```

```python
# map household types
ruralurban_map = {'1':'rural','2':'urban'}
df_bihar_demo['rural_urban'] = df_bihar_demo['Sector'].map(ruralurban_map)
df_bihar_demo.head()
df_bihar_demo= df_bihar_demo.drop(['Sector'], axis=1)
df_bihar_demo.head(100)
```

```python
# map religion
religion_map = {'1':'Hinduism','2':'Islam','3':'Christianity','4':'Sikhism','5':'Jainism','6':'Buddhism','7':'Zoroastrianism','9':'Others'}
df_bihar_demo['Religion_map'] = df_bihar_demo['Religion'].map(religion_map)
df_bihar_demo= df_bihar_demo.drop(['Religion'], axis=1)
df_bihar_demo.rename({'Religion_map': 'Religion'}, axis=1, inplace=True)
df_bihar_demo.head()
```

```python
df_bihar_demo.to_csv('hh_demograph_bihar.csv', index=True)
```
